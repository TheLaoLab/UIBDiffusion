## Backdoor Attack

You can train and test your own backdoored model based on running the code below, in which you can choose the dataset you want to use, different backdoor attack triggers and different backdoor attack targets.

This repo is running based on CIFAR10 dataset and the inperceptible trigger generated by ourselves named UAP_NOISE image. We chose HAT as the backdoor attack target.

### Training and testing

```python
!python UIBDiffusion.py --project default --mode train+measure --dataset CIFAR10 --batch 128 --epoch 50 --poison_rate 0.1 --trigger UAP_NOISE --target HAT --ckpt DDPM-CIFAR10-32 --fclip o -o --gpu 0
```

### Generate the clean samples and backdoor targets from a backdoored DM

```python
!python UIBDiffusion.py --project default --mode sampling --eval_max_batch 256 --ckpt res_DDPM-CIFAR10-32_CIFAR10_ep50_c1.0_p0.1_UAP_NOISE-HAT --fclip o --gpu 0
```

### Measure your backdoored model

If you only want to test the performance of your backdoored model, please use the code below, in which you should change the path : res_DDPM-CIFAR10-32_CIFAR10_ep50_c1.0_p0.1_UAP_NOISE-HAT into your own saved model's path.

To measure other models, for example, the clean models downloaded from Huggingface, you can directly use the UIBDiffusion.py script, but pay attention to changing the model's path accordingly.

```python
!python UIBDiffusion.py --project default --mode measure --ckpt res_DDPM-CIFAR10-32_CIFAR10_ep50_c1.0_p0.1_UAP_NOISE-HAT --fclip o --gpu 0
```
